```python
finance.rename(columns={'사업자번호':'BIZ_NO'}, inplace=True)
finance.dtypes
```

```python
finance.isnull().sum()
```

```python
activeall = pd.merge(activecop, finance, on='BIZ_NO')
print(activeall.shape)
activeall.head()
```

```python
# 사용할 칼럼만 불러오기
# 사업자번호, 기업명, 기업규모, 공기업, 본점지점구분, 국외투자법인여부, 설립일자, 상장일자,
# 홈페이지url, 직원수

activeall = activeall[['BIZ_NO', 'CMP_NM1', 'CMP_SCL', 'PBCO_GB', 'HDOF_BR_GB', 'FR_IVST_CORP_YN', 
                      'ESTB_DATE', 'LIST_DATE', 'HOMEPAGE_URL', 'EMP_CNT',
                      '결산년월', '매출채권', '자산총계', '부  채  총  계', '자본금', '자본총계',
                      '매출액', '영업이익（손실）', '당기순이익(손실)', '기업순이익률(%)',
                      '매출액총이익률(%)', '매출액영업이익률(%)', '매출액순이익률(%)',
                      '금융비용대매출액비율(%', '금융비용대부채비율(%)', '부채비율(%)',
                      '자기자본비율(%)', '유동부채비율(%)', '비유동부채비율(%)', '부채총계대 매출액(%)',
                      '미수금', '매출원가' ]]
print(activeall.shape)
activeall.head()
```
